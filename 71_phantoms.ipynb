```python
import io
df = pd.read_excel(io.BytesIO(uploaded['test data.xlsx']))
print(df)
```

```python
df.columns
```

```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
labels = df['Crop             ']
production_no = df['Production ']
yield_no = df['Yield ']

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(25,10))
rects1 = ax.bar(x-width/2, production_no, width, label = 'Production')
rects2 = ax.bar(x+width/2, yield_no, width, label= 'Yield')

ax.set_ylabel('Crops')
ax.set_xlabel('Yield ')
ax.set_title('Yield of crops in India')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation = 45)
ax.legend(fontsize=20)

def autolabel(rects):
  for rect in rects:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
    xy = (rect.get_x() + rect.get_width() / 2, height),
    xytext = (0, 1),
    textcoords = "offset points",
    ha='center', va='bottom')
    
autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```
